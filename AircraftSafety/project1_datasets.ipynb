```python
# Import packages
import pandas as pd
```

```python
summary_df = pd.read_csv('airline-safety.csv')
summary_df.head()
```

```python
new_df1 = pd.DataFrame(columns=['airline', 'type', 'incidents', 'fatalities', 'fatal_accidents'])
new_df1['airline'] = summary_df['airline']
new_df1['type'] = '85-99'
new_df1['incidents'] = summary_df['incidents_85_99']
new_df1['fatalities'] = summary_df['fatalities_85_99']
new_df1['fatal_accidents'] = summary_df['fatal_accidents_85_99']
new_df1.head()
```

```python
new_df2 = pd.DataFrame(columns=['airline', 'type', 'incidents', 'fatalities', 'fatal_accidents'])
new_df2['airline'] = summary_df['airline']
new_df2['type'] = '00-14'
new_df2['incidents'] = summary_df['incidents_00_14']
new_df2['fatalities'] = summary_df['fatalities_00_14']
new_df2['fatal_accidents'] = summary_df['fatal_accidents_00_14']
new_df2.head()
```

```python
combined_df = pd.concat([new_df1, new_df2])
combined_df.head()
```

```python
combined_df.tail()
```

```python
combined_df.to_csv('combined-airline-safety.csv')
```

```python
# Get vehicle crash data for comparison
crash_df_2015 = pd.read_csv('FARS2015NationalCSV/ACCIDENT.csv')
crash_df_2016 = pd.read_csv('FARS2016NationalCSV/ACCIDENT.csv')
crash_df_2017 = pd.read_csv('FARS2017NationalCSV/ACCIDENT.csv')
crash_df_2018 = pd.read_csv('FARS2018NationalCSV/ACCIDENT.csv')
#crash_df_2019 = pd.read_csv('FARS2019NationalCSV/accident_FARS.csv')
crash_df = pd.concat([crash_df_2015, crash_df_2016, crash_df_2017, crash_df_2018])
crash_df.head()
```

```python
crash_df['month-year'] = str(crash_df['MONTH']) +' '+ str(crash_df['YEAR'])
```

```python
crash_df.to_csv('car_crashes.csv')
```
