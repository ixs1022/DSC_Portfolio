```python
# Import packages
import pandas as pd
import numpy as np
```

```python
# Load Data files
airline_df = pd.read_excel('table_02_09_111020.xlsx', header=1, skipfooter=18)
airline_df.head()
```

```python
vehicle_df = pd.read_excel('Crash_Deaths_1975_2019.xlsx')
vehicle_df.head()
```

```python
# Transpose the airline data so it is organized by years (years as a column header, field in a row)
airline_transpose_df = airline_df.set_index('Unnamed: 0').T.rename_axis('Year').reset_index()

# Replace N with NaN
airline_transpose_df.replace(to_replace='N', value=np.NaN, inplace=True)
airline_transpose_df.head()
airline_transpose_df.tail()
```

```python
# Export transpose airlines file as csv
airline_transpose_df.to_csv('AirlineCrashes_1960_2019.csv')
```
